```python
def execute_program(prog):
    # Initialise register value
    reg = 1
    
    # Initialise clock
    cycle = 1
    
    cycle_checkpoints = list(range(20, 221, 40))
    
    sum_signal_strengths = 0
    
    for instruction_string in prog:
        #print(f"\nstart of cycle {cycle}")
        if instruction_string=="noop":
            instruction = instruction_string
            #print(f"instruction: {instruction}")
            #print(f"X = {reg}")
            if cycle in cycle_checkpoints:
                sum_signal_strengths += cycle*reg
                #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")            
            cycle += 1

        else:
            instr_split = instruction_string.split(" ")
            instruction = instr_split[0]
            value = int(instr_split[1])
            if instruction == "addx":
                #print(f"instruction: {instruction} (value: {value})")                
                #print(f"cycle: {cycle}")
                #print(f"X = {reg}")
                if cycle in cycle_checkpoints:
                    sum_signal_strengths += cycle*reg   
                    #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")                    
                cycle += 1
                #print(f"cycle: {cycle}")
                #print(f"X = {reg}") 
                if cycle in cycle_checkpoints:
                    sum_signal_strengths += cycle*reg   
                    #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")
                    
                reg += value
                cycle += 1
        #print(f"finished executing {instruction} at end of cycle {cycle-1}, X = {reg}")
                
    return sum_signal_strengths

execute_program(program_input)
```

```python
def during_cycle(cycle_number, crt_rows_list, sprite_pos):
    
    if (cycle_number-1) % 40 == 0:
        crt_r = ""
    else:
        crt_r = crt_rows_list[len(crt_rows_list)-1]
    
    print(f"During cycle\t{cycle_number}: CRT draws pixel in position {cycle_number-1}")
    if (cycle_number-1) % 40 in [sprite_pos-1, sprite_pos, sprite_pos+1]:
        crt_r += "#"
    else:
        crt_r += "."
    print(f"Current CRT row: \t{crt_r}\n")
    
    if cycle_number % 40 == 0:
        crt_rows_list.append(crt_r)        
    else:
        crt_rows_list[len(crt_rows_list)-1] = crt_r
        
    return crt_rows_list
    
def start_cycle(cycle_number, instr):
    print(f"\nStart cycle\t{cycle_number}: begin executing {instr}")
    
def end_cycle(cycle_number, instr, regx):
    print(f"End of cycle \t{cycle_number}: finish executing {instr} (Register X is now {regx})")
        

def execute_program_part2(prog):
    # Initialise register value
    reg = 1
    
    # Initialise sprite position
    sprite_position = 1
    
    # Initialise clock
    cycle = 1
    
    cycle_checkpoints = list(range(20, 221, 40))
    
    crt_rows = [""]
    
    #print(f"Sprite position: \t{sprite_position}")
        
    for instruction_string in prog:
        #print(f"\nstart of cycle {cycle}")
        start_cycle(cycle, instruction_string)        
        if instruction_string=="noop":
            instruction = instruction_string
            #print(f"instruction: {instruction}")
            #print(f"X = {reg}")
            crt_rows = during_cycle(cycle, crt_rows, reg)
                #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")    

        else:
            instr_split = instruction_string.split(" ")
            instruction = instr_split[0]
            value = int(instr_split[1])
            if instruction == "addx":
                #print(f"instruction: {instruction} (value: {value})")                
                #print(f"cycle: {cycle}")
                #print(f"X = {reg}")
                crt_rows = during_cycle(cycle, crt_rows, reg)
                    #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")                    
                cycle += 1
                #print(f"cycle: {cycle}")
                #print(f"X = {reg}") 
                crt_rows = during_cycle(cycle, crt_rows, reg)  
                    #print(f"\t --- signal strength: {cycle}*{reg} = {cycle*reg}")
                    
                reg += value
                #print(f"Sprite position: \t{reg}")
        end_cycle(cycle, instruction_string, reg)    
        cycle += 1
        #print(f"finished executing {instruction} at end of cycle {cycle-1}, X = {reg}")
                
    return crt_rows

crt = execute_program_part2(program_input)
display(crt)
```
